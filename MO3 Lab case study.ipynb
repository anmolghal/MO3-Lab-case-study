```python
# Define the superclass Vehicle
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

# Define the subclass Automobile, inheriting from Vehicle
class Automobile(Vehicle):
    def __init__(self, year, make, model, doors, roof):
        super().__init__("car")  # Set vehicle type to "car"
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof

    def display_info(self):
        print(f"Vehicle type: {self.vehicle_type}")
        print(f"Year: {self.year}")
        print(f"Make: {self.make}")
        print(f"Model: {self.model}")
        print(f"Number of doors: {self.doors}")
        print(f"Type of roof: {self.roof}")

# Main function to run the app
def main():
    print("Please enter the following details for the car:")
    
    year = input("Year: ")
    make = input("Make: ")
    model = input("Model: ")
    
    while True:
        try:
            doors = int(input("Number of doors (2 or 4): "))
            if doors in [2, 4]:
                break
            else:
                print("Please enter a valid number of doors (2 or 4).")
        except ValueError:
            print("Please enter a numeric value for the number of doors.")
    
    while True:
        roof = input("Type of roof (solid or sun roof): ").lower()
        if roof in ["solid", "sun roof"]:
            break
        else:
            print("Please enter a valid roof type (solid or sun roof).")
    
    # Create an instance of Automobile
    car = Automobile(year, make, model, doors, roof)
    
    # Display the information
    print("\nHere is the information you entered:")
    car.display_info()

# Run the main function
if __name__ == "__main__":
    main()
```
